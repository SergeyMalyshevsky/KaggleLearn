```python
import pandas as pd
import matplotlib
```

```python
train = pd.read_csv('./input/train.csv')
test = pd.read_csv('./input/test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train.Sex = train.Sex.replace('male', 0)
train.Sex = train.Sex.replace('female', 1)

test.Sex = test.Sex.replace('male', 0)
test.Sex = test.Sex.replace('female', 1)
```

```python
train.head()
```

```python
test.head()
```

```python
# Parameter inplace=True set default value for current DataFrame
train.fillna(train.mean(), inplace=True)
test.fillna(test.mean(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn import metrics
```

```python
# Create y
y = train.Survived

# Create X
features = ["Pclass", "Age", "Fare", "Sex"]
X = train[features]

# Split data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

# Create model
model = RandomForestClassifier(random_state=1)

# Fit model
model.fit(train_X, train_y)
```

![title](../images/precision-recall.png)

```python
preds_val = model.predict(val_X)
print(metrics.classification_report(val_y, preds_val))
```

```python
prediction = model.predict(test[features])
my_prediction = pd.DataFrame(prediction, test.PassengerId, columns = ["Survived"])
my_prediction.to_csv("./out/my_prediction.csv", index_label = ["PassengerId"])
```
